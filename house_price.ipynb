```python
import pandas as pd
train_df = pd.read_csv("house/train.csv", encoding="utf-8")
test_df = pd.read_csv("house/test.csv", encoding="utf-8")
train_df
```

```python
total_df = pd.concat([train_df, test_df], axis=0)
total_df
```

```python
data = total_df.drop(["SalePrice"], axis=1)
y_train = train_df["SalePrice"]
```

```python
# 補數值
med = data.median().drop(["MSSubClass"])
data = data.fillna(med)
```

```python
# one-hot
data = pd.get_dummies(data)
data = pd.get_dummies(data, columns=["MSSubClass"])
data = data.drop(["Id"], axis=1)
data
```

```python
s = data.isna().sum()
s[s>0]
```

```python
# 先補完 最後才用, 分開x_train , x_test
train_df.shape[0]
# loc: 列標籤, iloc: 0,1,2...
# iloc -> [""]
x_train = data.iloc[:train_df.shape[0]]
x_test = data.iloc[train_df.shape[0]:]
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
reg = RandomForestRegressor()
params = {
    "n_estimators":range(70, 75, 2),
    "max_depth":range(8,10)
}
cv = GridSearchCV(reg, params, scoring="r2", n_jobs=4, cv=10)
cv.fit(x_train, y_train)
print(cv.best_params_)
print(cv.best_score_)
```

```python
reg = RandomForestRegressor(n_estimators=72, max_depth=9)
reg.fit(x_train,y_train)
pre = reg.predict(x_test)
testid = test_df["Id"]
result_df = pd.DataFrame({
    "Id":testid,
    "SalePrice":pre
})
result_df.to_csv("rdf.csv", index=False, encoding="utf-8")
result_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
params = {
    "n_neighbors":range(5, 50)
}
cv = GridSearchCV(clf, params, cv=10, n_jobs=4)
cv.fit(x_train_scale, y_train)
print(cv.best_params_)
print(cv.best_score_)
```

```python
pre = cv.best_estimator_.predict(x_test_scale)
result_df = pd.DataFrame({
    "Id":testid,
    "SalePrice":pre
})
result_df.to_csv("knn2.csv", index=False, encoding="utf-8")
result_df
```
